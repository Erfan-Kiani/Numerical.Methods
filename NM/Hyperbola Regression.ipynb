```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

def hyperbola(x, a, b):
    return a / x + b

x = np.array([-8, -6, -4])
y = np.array([30, 10, 9])

popt, pcov = curve_fit(hyperbola, x, y)

print("Coefficients: ", popt)

plt.plot(x, y, 'o', label='data')
plt.plot(x, hyperbola(x, *popt), 'r-', label='fit')
plt.legend()
plt.show()

```
