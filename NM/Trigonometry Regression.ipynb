```python
#Trigonometry Regression 

import numpy as np
import matplotlib.pyplot as plt

x = ([3.6, 4, 4.4, 4.8, 5.2, 5.6, 6])
y = ([2.0, 2.9, 3.5, 4.0, 4.2, 4, 3.5])

X = np.column_stack((np.ones(7), np.sin(x), np.cos(x)))
model = np.linalg.lstsq(X, y, rcond=None)[0]

a = model[0]
b = model[1]
c = model[2]

print("a:", a)
print("b:", b)
print("c:", c)

plt.plot(x, y, 'o', label='Original data', markersize=3)
plt.plot(x, a + b*np.sin(x) + c*np.cos(x), 'r', label='Fitted curve')
plt.legend()
plt.show()
```
