```python
# Fourth-Order Runge-Kutta Method

import numpy as np
import matplotlib.pyplot as plt

def f(x, y):
    return x+y

def rk4(f, x0, y0, x1, n):
    h = (x1 - x0) / float(n)
    xi = x0
    yi = y0
    for i in range(n):
        k1 = h * f(xi, yi)
        k2 = h * f(xi + 0.5*h, yi + 0.5*k1)
        k3 = h * f(xi + 0.5*h, yi + 0.5*k2)
        k4 = h * f(xi + h, yi + k3)
        yi += (k1 + 2*k2 + 2*k3 + k4) / 6.0
        xi += h
    return yi
print("Answer FORKM is:", rk4(f, 0, 1, 0.1,1 ))

x0 = 0
y0 = 1
x1 = 0.1
n = 1

x = np.linspace(x0, x1, n+1)
y = np.zeros(n+1)
y[0] = y0

for i in range(n):
    y[i+1] = rk4(f, x[i], y[i], x[i+1], 1)

plt.plot(x,y,'o-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()

# Second-Order Runge-Kutta Method

def f(x, y):
    return x+y

def rk4(f, x0, y0, x1, n):
    h = (x1 - x0) / float(n)
    xi = x0
    yi = y0
    for i in range(n):
        k1 = h * f(xi, yi)
        k2 = h * f(xi + h, yi + k1)
        yi += (k1 + k2) /2.0
        xi += h
    return yi
print("Answer FORKM is:", rk4(f, 0, 1, 0.1,2 ))

x0 = 0
y0 = 1
x1 = 0.1
n = 

x = np.linspace(x0, x1, n+1)
y = np.zeros(n+1)
y[0] = y0

for i in range(n):
    y[i+1] = rk4(f, x[i], y[i], x[i+1], 1)

plt.plot(x,y,'o-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
