```python
#Exponential Regression 

import numpy as np
import matplotlib.pyplot as plt

x = np.array([1, 2, 3, 4])
y = np.array([7, 11, 17, 27])

coefficients = np.polyfit(x, np.log(y), 1)
a = np.exp(coefficients[1])
b = coefficients[0]

print("a:", a)
print("b:", b)

plt.plot(x, y, 'o', label='Original data', markersize=3)
plt.plot(x, a*np.exp(b*x), 'r', label='Fitted curve')
plt.legend()
plt.show()
```
