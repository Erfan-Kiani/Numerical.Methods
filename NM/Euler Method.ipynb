```python
# Euler Method 

import numpy as np
import matplotlib.pyplot as plt


def euler(f, x0, y0, h, n):
    x = np.zeros(n+1)
    y = np.zeros(n+1)
    x[0] = x0
    y[0] = y0
    for i in range(n):
        y[i+1] = y[i] + h * f(x[i], y[i])
        x[i+1] = x[i] + h
    return x, y

def f(x, y):
    return 1 - (y/x)

x0 = 2
y0 = 2
h = 0.1
n = 1

x_euler, y_euler = euler(f, x0, y0, h, n)

print("The final answer is:",y_euler[-1])

t_exact = np.linspace(0,n*h,n+1)
y_exact = np.exp(-t_exact)

plt.plot(t_exact,y_exact,'r',label='Exact')
plt.plot(x_euler,y_euler,'b',label='Euler')
plt.legend()
plt.show()
```
